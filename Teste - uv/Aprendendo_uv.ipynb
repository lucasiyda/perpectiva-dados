# Atalhos úteis

1. Ctrl + Shift + T: Abre o terminal
2. Ctrl + Shift + V: Cola coisas no terminal
3. Ctrl + H: Mostra itens ocultos
4. Ctrl + L: Limpa o terminal


# Comandos básicos do git

1. `git clone`: Clona o repositório do git hub (*Exemplo: git clone https://github.com/lucasiyda/perpectiva-dados.git*)
   
2. `git status`: Mostra o status se algum arquivo foi alterado ou tem algum arquivo (*Exemplo: Usamos para verificar se algum arquivo que alteramos foi realmente alterado*)

4. `git add`: Adiciona algum arquivo novo para comitta (*Exemplo: git add README.md*)

5. `git diff`: Mostra as mudanças realizadas (*Exemplo: Vai nos informar quais foram as mudanças realizadas no README.md*)

6. `git commit -m`: Comentar o que foi alterado, para melhor entendimento das etapas realizadas (*Exemplo: git commit -m "Exluir arquivos inúteis*)

7. `git push`: Adicionamos as alterações ao diretório do Git Hub (*Exemplo: Quando queremos atualizar as informações realizada na máquina e enviar para o repositório do git hub*)

8. `git log`: Mostra todas as alterações feitas e commits (*Exemplo: Saber quais foram todas alterações realizadas e quando foram*)

9. `git restore`: Restaura mudanças dentro dos arquivos (*Exemplo: Arquivo Texto.txt foi alterado mas você quer desfazer as alterações sem ir dentro do arquivo e apagar, então da um git restore Texto.txt e as mudanças voltaram ao que estava*)

10. `git restore --staged`: Restaura arquivos que foram adicionados para commitar (*Exemplo: Adicionei um arquivo novo e fiz git add Arquivo_novo.txt, porém você não quer commitar esse arquivo, então basta realizar git restore --staged Arquivo_novo.txt*)

# Comandos básicos do terminal

1. `cd`: "Change Directory" serve para abrir o terminal no diretório correto (*Exemplo: cd Pasta_correta*)
2. `ls`: Lista todos os arquivos que possuem naquele diretório
3. `ls -a`: Lista todos os arquivos que possuem naquele diretório mais os que estão ocultos
4. `touch`: Cria um arquivo na pasta diretamente do terminal (*Exemplo: touch texto.txt*)
5. `cat`: Mostra tudo que está dentro do arquivo, serve para ver o que está dentro do arquivo sem precisar abrir ele (*Exemplo: cat texto.txt*)
6. `mkdir`: Cria uma pasta nova no diretório
