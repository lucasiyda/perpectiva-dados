# Comandos básicos do git

1. `git clone`: Clona o repositório do git hub (*Exemplo: git clone https://github.com/lucasiyda/perpectiva-dados.git*)
   
2. `git status`: Mostra o status se algum arquivo foi alterado ou tem algum arquivo (*Exemplo: Usamos para verificar se algum arquivo que alteramos foi realmente alterado*)

4. `git add`: Adiciona algum arquivo novo para comitta (*Exemplo: git add README.md*)

5. `git diff`: Mostra as mudanças realizadas (*Exemplo: Vai nos informar quais foram as mudanças realizadas no README.md*)

6. `git commit -m`: Comentar o que foi alterado, para melhor entendimento das etapas realizadas (*Exemplo: git commit -m "Exluir arquivos inúteis*)

7. `git push`: Adicionamos as alterações ao diretório do Git Hub (*Exemplo: Quando queremos atualizar as informações realizada na máquina e enviar para o repositório do git hub*)

8. `git log`: Mostra todas as alterações feitas e commits (*Exemplo: Saber quais foram todas alterações realizadas e quando foram*)


# Comandos básicos do uv

Antes de tudo devemos baixar o pacote do ub em "*wget -qO- https://astral.sh/uv/install.sh | sh*"

1. `uv add`: Instala pacote novo (*"Exemplo: uv add jupyter notebook"*)
2. `uv run`: Cria um notebook na Web (*"Exemplo: uv run jupyter notebook"*)
