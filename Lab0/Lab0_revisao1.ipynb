```python
# Passo 1
# Dicionário
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}

```

```python
# Passo 2
# Importar Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Passo 3
# Gerar Amostras das Distribuições
normal = np.random.normal(5,1, size = 100)
exponencial = np.random.exponential(4, size = 100)
uniforme = np.random.uniform(0,5, size = 100)
```

```python
# Passo 4
# Calcular Estatísticas Básicas

# Normal
media_normal = np.mean(normal) # Média
variancia_normal = np.var(normal) # Variancia
desvio_padrao_normal = np.sqrt(variancia_normal) # Desvio Padrao
params_normal = [5,1] # Parametros

# Exponencial
media_exponencial = np.mean(exponencial) # Média
variancia_exponencial = np.var(exponencial) # Variancia
desvio_padrao_exponencial = np.sqrt(variancia_exponencial)  # Desvio Padrao
params_exponencial = 4 # Parametros

# Uniforme
media_uniforme = np.mean(uniforme) # Média
variancia_uniforme = np.var(uniforme) # Variancia
desvio_padrao_uniforme = np.sqrt(variancia_uniforme)  # Desvio Padrao
params_uniforme = [0,5] # Parametros
```

```python
# Passo 4.1
# Criar "chave" para as Estatísticas

distribuicoes["Normal"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None}
distribuicoes["Exponencial"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None}
distribuicoes["Uniforme"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None}
```

```python
# Passo 4.2
# Adicionar Informações no Dicionário

# Normal
distribuicoes["Normal"]["dados"] = normal
distribuicoes["Normal"]["estatisticas"]["Media"] = media_normal
distribuicoes["Normal"]["estatisticas"]["Variancia"] = variancia_normal
distribuicoes["Normal"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_normal
distribuicoes["Normal"]["params"]["loc"] = params_normal[0]
distribuicoes["Normal"]["params"]["scale"] = params_normal[1]

# Exponencial
distribuicoes["Exponencial"]["dados"] = exponencial
distribuicoes["Exponencial"]["estatisticas"]["Media"] = media_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Variancia"] = variancia_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_exponencial
distribuicoes["Exponencial"]["params"]["scale"] = params_exponencial

# Uniforme
distribuicoes["Uniforme"]["dados"] = uniforme
distribuicoes["Uniforme"]["estatisticas"]["Media"] = media_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Variancia"] = variancia_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_uniforme
distribuicoes["Uniforme"]["params"]["low"] = params_uniforme[0]
distribuicoes["Uniforme"]["params"]["high"] = params_uniforme[1]
```

```python
# Passo 5
# Histogramas

# Normal
plt.subplot(1, 3, 1) # Plotar um do lado do outro
plt.hist(normal)
plt.title("Normal")

# Exponencial
plt.subplot(1, 3, 2)
plt.hist(exponencial)
plt.title("Exponencial")

# Uniforme
plt.subplot(1, 3, 3)
plt.hist(uniforme)
plt.title("Uniforme")

plt.tight_layout() # Ajusta automaticamente os espaços entre cada gráfico
```

```python
# Passo 6
# Organizar resultados em DataFrame


pd.DataFrame({
    "Normal": distribuicoes["Normal"]["estatisticas"],
    "Exponencial": distribuicoes["Exponencial"]["estatisticas"],
    "Uniforme": distribuicoes["Uniforme"]["estatisticas"]
})

```

```python
# Passo 7

# Adicionar nova distribuição Poisson no Dicionário
distribuicoes["Poisson"] = {"params": {"lam": 2}, "dados": None, "estatisticas": None}
distribuicoes["Poisson"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None}

# Gerar dados
poisson = np.random.poisson(6, size = 100)

# Calcular Estatísticas Básicas
media_poisson = np.mean(poisson)
variancia_poisson = np.var(poisson) # Variancia
desvio_padrao_poisson = np.sqrt(variancia_poisson)  # Desvio Padrao
params_poisson = 6 # Parametros

# Adicionar Informações no Dicionário
distribuicoes["Poisson"]["dados"] = poisson
distribuicoes["Poisson"]["estatisticas"]["Media"] = media_poisson
distribuicoes["Poisson"]["estatisticas"]["Variancia"] = variancia_poisson
distribuicoes["Poisson"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_poisson
distribuicoes["Poisson"]["params"]["lam"] = params_poisson

```

```python
from IPython.core.interactiveshell import dis
# Passo 8
# Estatísticas adicionais

# Normal
min_normal = np.min(normal) # Minimo
max_normal = np.max(normal) # Maximo
quartis_normal = np.quantile(normal,[0.25,0.5,0.75]) # Primeiro, Segundo e Terceiro Quartis

# Exponencial
min_exponencial = np.min(exponencial) # Minimo
max_exponencial = np.max(exponencial) # Maximo
quartis_exponencial = np.quantile(exponencial,[0.25,0.5,0.75])

# Uniforme
min_uniforme = np.min(uniforme) # Minimo
max_uniforme = np.max(uniforme) # Maximo
quartis_uniforme = np.quantile(uniforme,[0.25,0.5,0.75]) # Primeiro, Segundo e Terceiro Quartis

# Poisson
min_poisson = np.min(poisson) # Minimo
max_poisson = np.max(poisson) # Maximo
quartis_poisson = np.quantile(poisson,[0.25,0.5,0.75]) # Primeiro, Segundo e Terceiro Quartis

# Atualizar o campo Estatísticas no Dicionário
distribuicoes["Normal"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None, "Minimo": None, "Maximo": None, "Primeiro Quartil": None, "Mediana": None, "Terceiro Quartil": None}
distribuicoes["Exponencial"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None, "Minimo": None, "Maximo": None, "Primeiro Quartil": None, "Mediana": None, "Terceiro Quartil": None}
distribuicoes["Uniforme"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None, "Minimo": None, "Maximo": None, "Primeiro Quartil": None, "Mediana": None, "Terceiro Quartil": None}
distribuicoes["Poisson"]["estatisticas"] = {"Media": None, "Variancia": None, "Desvio Padrao": None, "Minimo": None, "Maximo": None, "Primeiro Quartil": None, "Mediana": None, "Terceiro Quartil": None}

# Adicionar as Informações no Dicionário
# Normal
distribuicoes["Normal"]["estatisticas"]["Media"] = media_normal
distribuicoes["Normal"]["estatisticas"]["Variancia"] = variancia_normal
distribuicoes["Normal"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_normal
distribuicoes["Normal"]["estatisticas"]["Minimo"] = min_normal
distribuicoes["Normal"]["estatisticas"]["Maximo"] = max_normal
distribuicoes["Normal"]["estatisticas"]["Primeiro Quartil"] = quartis_normal[0]
distribuicoes["Normal"]["estatisticas"]["Mediana"] = quartis_normal[1]
distribuicoes["Normal"]["estatisticas"]["Terceiro Quartil"] = quartis_normal[2]

# Exponencial
distribuicoes["Exponencial"]["estatisticas"]["Media"] = media_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Variancia"] = variancia_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Minimo"] = min_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Maximo"] = max_exponencial
distribuicoes["Exponencial"]["estatisticas"]["Primeiro Quartil"] = quartis_exponencial[0]
distribuicoes["Exponencial"]["estatisticas"]["Mediana"] = quartis_exponencial [1]
distribuicoes["Exponencial"]["estatisticas"]["Terceiro Quartil"] = quartis_exponencial[2]

# Uniforme
distribuicoes["Uniforme"]["estatisticas"]["Media"] = media_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Variancia"] = variancia_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Minimo"] = min_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Maximo"] = max_uniforme
distribuicoes["Uniforme"]["estatisticas"]["Primeiro Quartil"] = quartis_uniforme[0]
distribuicoes["Uniforme"]["estatisticas"]["Mediana"] = quartis_uniforme [1]
distribuicoes["Uniforme"]["estatisticas"]["Terceiro Quartil"] = quartis_uniforme[2]

# Poisson
distribuicoes["Poisson"]["estatisticas"]["Media"] = media_poisson
distribuicoes["Poisson"]["estatisticas"]["Variancia"] = variancia_poisson
distribuicoes["Poisson"]["estatisticas"]["Desvio Padrao"] = desvio_padrao_poisson
distribuicoes["Poisson"]["estatisticas"]["Minimo"] = min_poisson
distribuicoes["Poisson"]["estatisticas"]["Maximo"] = max_poisson
distribuicoes["Poisson"]["estatisticas"]["Primeiro Quartil"] = quartis_poisson[0]
distribuicoes["Poisson"]["estatisticas"]["Mediana"] = quartis_poisson [1]
distribuicoes["Poisson"]["estatisticas"]["Terceiro Quartil"] = quartis_poisson[2]

```

```python
# Passo 8
# DataFrame
pd.DataFrame({
    "Normal": distribuicoes["Normal"]["estatisticas"],
    "Exponencial": distribuicoes["Exponencial"]["estatisticas"],
    "Uniforme": distribuicoes["Uniforme"]["estatisticas"],
    "Poisson": distribuicoes["Poisson"]["estatisticas"]
})
```
