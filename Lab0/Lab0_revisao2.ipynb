```python
# Numpy
# Exercicio 1 - Criar vetor com 20 números aleatorios entre 0 e 100

np.random.seed(144) # Colocar semente

vetor = np.random.randint(0, 100, 20)
print(f"Vetor: {vetor}")


# Media
media = np.mean(vetor)
print(f"Média: {media:.2f}")

# Desvio Padrao
variancia = np.var(vetor, ddof=1) # Variancia amostral
desvio_padrao = np.sqrt(variancia)
print(f"Desvio Padrao: {desvio_padrao:.2f}")


# Valor Máximo
maximo = np.max(vetor)
print(f"Maximo: {maximo:.2f}")
```

```python
# Numpy
# Exercicio 2 - Gere uma matriz 5x5 com valores da distribuição normal padrão.

np.random.seed(144) # Colocar semente

# Matriz
matriz = np.random.normal(0,5, size=(5, 5))
print(f"Matriz: {matriz}")

# Calcule a soma de cada linha
for i in range(len(matriz)):
  soma_linha = np.sum(matriz[i])
  print(f"Soma da linha {i+1}: {soma_linha}")

# Calcule a soma de cada coluna
for j in range(len(matriz)):
  soma_coluna = np.sum(matriz[:,j])
  print(f"Soma da coluna {j+1}: {soma_coluna}")

# Extraia a diagonal principal
diag_principal = np.diag(matriz)
print(f"Diagonal Principal: {diag_principal}")
```

```python
# Numpy
# Exercicio 3 - Implemente uma função que recebe dois vetores NumPy e retorna:
## Produto escalar (np.dot).
## Distância euclidiana entre eles (np.linalg.norm).

def calcula (vetor1, vetor2):
  prod_escalar = np.dot(vetor1, vetor2)
  dist_euclidiana = np.linalg.norm(vetor1 - vetor2)
  print(f"Produto Escalar: {prod_escalar}")
  print(f"Distancia Euclidiana: {dist_euclidiana}")

calcula(np.random.randint(0, 5, 10),np.random.randint(0, 5, 10))
```

```python
# Numpy
# Exercicio 4 - Crie uma matriz quadrada com valores inteiros aleatórios entre 5 e -5

np.random.seed(144) # Colocar semente

# Matriz
A =  np.random.randint(-5, 5, size = (4,4))
print(f"Matriz A : {A}")

# A é invertivel?
determinante = np.linalg.det(A) # Difrente de zero então é invertivel

# Calcular Inversa de A
A_inversa = np.linalg.inv(A)
print(f"Matriz A Inversa: {A_inversa}")

# Verificar se A*A^-1 = I
multiplica_matriz = np.matmul(A, A_inversa)
print(np.round(multiplica_matriz, 5)) # Deixa apenas 1 casa decimal
```

```python
# Numpy
# Exercicio 5 - Gere uma matriz simétrica 5x5 a partir de uma matriz aleatória definida por B = 1/2(M+M^T)

np.random.seed(144) # Colocar semente

# Matriz Aleatoria
M = np.random.randint(0, 10, size = (5,5))
print(f"Matriz M : {M}")

# Matriz B
B = 0.5*(M + M.T)
print(f"Matriz B : {B}")

# Calcula autovetores e autovalores
autovalores, autovetores = np.linalg.eig(B)
print(f"Autovalores: {autovalores}")
print(f"Autovetores: {autovetores}")

# Relaçao B*v = lambda*v
lado_esquerdo = np.matmul(B, autovetores)
lado_direito = autovalores *autovetores
print(f"lado_esquerdo: {lado_esquerdo}")
print(f"lado_direito: {lado_direito}")
```

```python
# Numpy
# Exercicio 6 - Considere um sistema linear Ax = y, com A R^3x3 e y om R^3

np.random.seed(144) # Colocar semente

# Gerar A e y
A = np.random.randint(0, 5, size = (3,3))
y = np.random.randint(0, 5, size = 3)
print(f"Matriz A: {A}")
print(f"Vetor y: {y}")

# Resolva o sistema
x = np.linalg.solve(A, y)
print(f"Vetor x: {x}")

# Verifique solução com np.linalg.norm
erro = np.linalg.norm(np.matmul(A, x) - y)
print(f"Erro: {erro}")
```

```python
# Numpy
# Exercicio 7 - Simule 100 vetores bidimensionais com distribuição normal padrão e forme a matriz de covariância empírica


np.random.seed(144) # Colocar semente

# Vetores
S = np.random.normal(0, 1, size=(100, 2))
print(f"Vetores:")
print(np.round(S, 2))

# Calcular Matriz de Covariancia Empírica
S_covariancia = np.cov(S, rowvar=False) # Porque temos apenas duas colunas ou seja 2 variaveis
print(f"Matriz Covariancia:")
print(np.round(S_covariancia, 3))

# Calcula autovetores e autovalores
autovalores, autovetores = np.linalg.eig(S_covariancia)
print(f"Autovalores: {autovalores}")
print(f"Autovetores: {autovetores}")


# Mostre que os autovetores correspondem às direções principais da variabilidade dos dados.
v1 = autovetores[:, 0] # Pegar principal autovetor (maior variância)

proj = np.matmul(S, v1) # Projetar os dados nessa direção

var_proj = np.var(proj) # Variância dos dados projetados
print(f"Maior autovalor:{autovalores[0]}")
print(f"Variancia na Direção do Autovetor  :{var_proj}")

# Como ambos são próximos mostramos que autovetores correspondem às direções principais da variabilidade dos dados
```

```python
# Pandas
# Exercicio 7

np.random.seed(144) # Colocar semente

# Produtos
produtos = ["Arroz", "Feijão", "Macarrão", "Carne", "Frango"]

# DataFrame
vendas = pd.DataFrame({
    "Produtos": np.random.choice(produtos, size=10),
    "Preço": np.random.uniform(5, 50, size=10).round(2),
    "Quantidade": np.random.randint(1, 10, size=10)
})

# Nova coluna receita
vendas["Receita"] = vendas["Preço"] * vendas["Quantidade"]
print(vendas)

# Agrupe por produto e traga receita total
receita_total = vendas.groupby("Produtos")["Receita"].sum()
print(receita_total)

# Ordene os mais vendidos
receita_total = receita_total.sort_values(ascending=False)
print(receita_total)
```
